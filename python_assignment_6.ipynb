```python
from urllib.request import urlopen
import pandas as pd
import json
page = urlopen('https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json')
data_json = json.loads(page.read())
df=pd.DataFrame.from_dict(data_json)

attributr=['id','num','name','img','type','height','weight','candy','egg','spawn_chance','avg_spawns','spawn_time','multipliers','weaknesses']
dic={}
for i in attributr:
  temp=[]
  for j in range(len(df)):
    temp.append(df.values[j][0][i])
  dic[i]=temp
  
data=pd.DataFrame(dic)
```

```python
data[data['avg_spawns']<5.00]['name']
```

# Get all Pokemons that have less than 4 weaknesses 

```python
index=[]
for i in range(len(data['weaknesses'])):
  if(len(data.iloc[i]['weaknesses'])<4):
    index.append(i)

print((data.iloc[index])['name'])
```

```python
data[data['multipliers'].isnull()]['name']
```

#Get all Pokemons whose spawn time is less than 300 seconds. 

```python
data['hour']=data['spawn_time'].str.split(":").str[0]
data['min']=data['spawn_time'].str.split(":").str[1]
```

```python
import numpy as np
```

```python
data['hour']=data['hour'].replace('N/A',0)
data['min']=data['min'].replace('N/A',0)
```

```python
data['hour']=data['hour'].astype(int)
data['min']=data['min'].astype(int)
```

```python
data['time_in_sec']=(data['hour']*60+data['min'])*60
```

```python
data[data['time_in_sec']<300]
```
