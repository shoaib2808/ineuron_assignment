# Get all the Earth meteorites that fell before the year 2000 

```python
df['year_exact']=df['year'].str.split("-").str[0]
```

```python
df['year_exact']=df['year_exact'].fillna(0000)
```

```python
df['year_exact']=df['year_exact'].astype(int)
```
